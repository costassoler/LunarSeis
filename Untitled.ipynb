```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import time

Data = np.load("recording2020_5_12_1_43.npz")
UnixStart = Data["times"][0]
millisCorrect = Data["times"][1]
millis = Data["times"][2:]
deltas = millis-millisCorrect
unixTimestamps = (deltas/1000)+UnixStart
print(millis)
times = [datetime.datetime.utcfromtimestamp(ts) for ts in unixTimestamps]
print(times)
signals = Data["signals"][2:]
```

```python
%matplotlib notebook
plt.plot(times,signals)
plt.ylim(-500,1300)
plt.xticks(rotation=25)
plt.show()
```

```python
signalFT = np.fft.fft(signals)
n = signals.size
freq=np.fft.fftfreq(n,0.025)
plt.plot(freq,signalFT)
```

```python
for i in range (0,n):
    if (freq[i]>1) or (freq[i]<-1):
        signalFT[i]=0
```

```python
signalFilt = np.fft.ifft(signalFT)
plt.plot(times,signals,'b-')
plt.plot(times,signalFilt,'r-')
plt.ylim(0,1023)
plt.xticks(rotation=25)
plt.show()
```

# Noise Process LowR Voltage Divider

```python
import numpy as np
import matplotlib.pyplot as plt
import datetime
import time

Data = np.load("StockRecording2020_5_13_19_37.npz")
UnixStart = Data["times"][0]
millisCorrect = Data["times"][1]
millis = Data["times"][2:]
deltas = millis-millisCorrect
unixTimestamps = (deltas/1000)+UnixStart
#print(millis)
times = [datetime.datetime.utcfromtimestamp(ts) for ts in unixTimestamps]
print(times)
signals = Data["signals"][2:]

%matplotlib notebook
plt.plot(times,signals)
plt.ylim(-500,1300)
plt.xticks(rotation=25)
plt.show()
```

```python
A=np.zeros((1,len(signals)))

valueP = signals[0]
print(valueP)
n=0
m=0
for i in range (0,len(signals)):
    value = signals[i]
    if (abs(value-valueP)>10):
        valueF = valueP
        A[0][i]=valueF
        n+=1
        print(n)
    else:
        valueF = value
        A[0][i]=value
        m+=1
        print(m)
    
    valueP = valueF

%matplotlib notebook
plt.plot(times,A[0])
plt.ylim(-500,1300)
plt.xticks(rotation=25)
plt.show()
```
